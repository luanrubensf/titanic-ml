```python
df['Fare'].plot.hist(bins=40)
# distribuição do custo com passagem
```

```python
df[df['Fare'] > 300]
```

```python
sns.barplot(x='Survived',y='Fare', data=df, palette='rainbow')
# a média de gasto de quem sobreviveu é maior que a de quem não sobreviveu
```

```python
sns.stripplot(x='Survived',y='Fare', hue='Sex', data=df, palette='rainbow',jitter=True)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
# indica o custo da passagem para sobrevivente ou não, dividido por sexo.
```

```python
sns.violinplot(x='Survived', y='Fare', data=df, palette='rainbow', width=0.9, inner='point')
# mostra que a grande maioria de não sobreviventes possuíam menor renda, embora existam casos isolados
```

```python
sns.boxplot(x="Survived", y="Fare",data=df, palette='rainbow')
# Distribuição do custo de passagem por sobrevivente/nao sobrevivente. É possível notar que 
# os sobreviventes geralmente gastaram mais
```

```python
sns.boxplot(x="Survived", y="Fare", hue='Sex', data=df, palette='rainbow')
# Incluindo divisão por sexo, a média de gasto das mulheres e homens que sobreviveram, é maior
```

```python
sns.boxplot(x="Sex", y="Fare", data=df, palette='rainbow')
# Média de gasto das mulheres é maior que a dos homens
```
