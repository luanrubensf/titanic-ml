```python
sns.barplot(x='Survived',y='Fare', data=df, palette='rainbow')
# a mdia de gasto de quem sobreviveu  maior que a de quem no sobreviveu
```

```python
sns.stripplot(x='Survived',y='Fare', hue='Sex', data=df, palette='rainbow',jitter=True)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
# indica o custo da passagem para sobrevivente ou no, dividido por sexo.
```

```python
sns.violinplot(x='Survived', y='Fare', data=df, palette='rainbow', width=0.9, inner='point')
# mostra que a grande maioria de no sobreviventes possuam menor renda, embora existam casos isolados
```

```python
sns.boxplot(x="Survived", y="Fare",data=df, palette='rainbow')
# Distribuio do custo de passagem por sobrevivente/nao sobrevivente.  possvel notar que 
# os sobreviventes geralmente gastaram mais
```

```python
sns.boxplot(x="Survived", y="Fare", hue='Sex', data=df, palette='rainbow')
# Incluindo diviso por sexo, a mdia de gasto das mulheres e homens que sobreviveram,  maior
```

```python
sns.boxplot(x="Sex", y="Fare", data=df, palette='rainbow')
# Mdia de gasto das mulheres  maior que a dos homens
```
